```python
import unicodedata
def remove_accents(s: str) -> str:
    if not isinstance(s, str):
        return s
    return (
        unicodedata.normalize("NFKD", s)
        .encode("ascii", "ignore")
        .decode("ascii")
    )

print(remove_accents("Jokić"))
```
